```python
# Sample notebook for generating image from the FastAPI server
# python main.py --config ./configs/{config}.json --host {x} --port {y}

import requests
import io

# cuz i'm saved by the bell

prompt = """
setting: High school hallway
background: Rows of lockers, posters, vibrant wall colors
time_of_day: Indoor, artificial lighting

character1:
  name: Zack Morris
  position: Center
  gender: Male
  age: Young adult
  hair: Style typical of 90s fashion
  facial_features: Animated, enthusiastic expression
  clothing: Casual 90s attire, bright colors or patterns
  accessories: Oversized 90s cell phone with extended antenna
  action: Talking enthusiastically on the cell phone

supporting_elements:
  extras: Background students in 90s fashion
  props: School supplies, backpacks

composition:
  framing: Medium shot
  perspective: Slightly low angle
  focus: Zack Morris in sharp focus

camera:
  angle: Typical sitcom framing
  shot_type: Single-character focus
  depth_of_field: Moderate, some background detail visible

visual_style:
  color_palette: Saturated primaries, pastels, neon accents
  lighting: Bright, even illumination, minimal shadows
  effects: Slightly soft focus, subtle film grain

technical_details:
  aspect_ratio: 4:3
  era: Early 1990s television sitcom
  show_reference: Saved by the Bell

mood: Energetic, comedic, youthful exuberance
"""

res = requests.post(
    "http://127.0.0.1:8088/generate",
    json={
        "width": 1024,
        "height": 1024,
        "num_steps": 25,
        "guidance": 3.5,
        "prompt": prompt,
        "seed": 1337,
    },
    stream=True,
)

# display image from response
from PIL import Image
from IPython.display import display

img = Image.open(io.BytesIO(res.content))
display(img)
```
